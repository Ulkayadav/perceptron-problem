```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
#this function is used to geneate a random classification dataset
X,y=make_classification(n_samples=100,n_features=2,n_classes=2,n_redundant=0,n_clusters_per_class=1,hypercube=False,random_state=41,class_sep=15,n_informative=1)
```

```python
y.shape
```

```python
plt.figure(figsize=(4,4))
plt.scatter(X[:,0],X[:,1],s=100,cmap="winter")
plt.show()
```

```python
def perceptron(X,y):
    lr=0.1
    w1=w2=b=1
    for j in range(1000):
        for i in range(X.shape[0]):
            z=w1*X[i][0]+w2*X[i][1]+b
        
            if z*y[i]<0:
                w1=w1+lr*X[i][0]*y[i]
                w2=w2+lr*X[i][1]*y[i]
                b=b+lr*y[i]
    return w1,w2,b
            
```

```python
w1,w2,b=perceptron(X,y)
```

```python
b
```

```python
m=-(w1/w2)
c=-(b/w2)
print("m",m)
print("c",c)
```

```python
x_input=np.linspace(-4,4,100)
y_input=m*x_input+c
plt.plot(x_input,y_input,color='red')
plt.scatter(X[:,0],X[:,1],s=100,cmap="winter",c=y)
plt.ylim(-3,3)
```
